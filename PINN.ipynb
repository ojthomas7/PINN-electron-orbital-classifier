```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import os
from PIL import Image
import matplotlib.pyplot as plt

def advanced_physics_informed_loss(y_true, y_pred, n_pred, l_pred, m_pred):
    """
    Enhanced physics-informed loss function with more sophisticated quantum mechanical constraints
    """
    # Standard categorical crossentropy with label smoothing
    standard_loss = tf.keras.losses.categorical_crossentropy(
        y_true, 
        y_pred, 
        label_smoothing=0.1
    )
    
    # Convert predictions to float32
    n_float = tf.cast(n_pred + 1, tf.float32)  # Add 1 since n starts from 1
    l_float = tf.cast(l_pred, tf.float32)
    m_float = tf.cast(m_pred, tf.float32)
    
    # Enhanced physics constraints
    
    # 1. l must be less than n (with smooth boundary)
    l_constraint = tf.reduce_mean(
        tf.nn.softplus(l_float - (n_float - 1))
    )
    
    # 2. |m| must be less than or equal to l (with smooth boundary)
    m_constraint = tf.reduce_mean(
        tf.nn.softplus(tf.abs(m_float) - l_float)
    )
    
    # 3. Energy level constraint (n affects energy most significantly)
    energy_constraint = tf.reduce_mean(
        tf.nn.softplus(-n_float)  # Ensure n is positive
    )
    
    # 4. Angular momentum conservation
    angular_momentum_constraint = tf.reduce_mean(
        tf.nn.softplus(l_float * (l_float + 1) - n_float * (n_float - 1))
    )
    
    # Combine all constraints with weighted importance
    total_loss = (
        standard_loss +
        5.0 * l_constraint +
        5.0 * m_constraint +
        2.0 * energy_constraint +
        3.0 * angular_momentum_constraint
    )
    
    return total_loss

def create_improved_pinn_model(input_shape, max_n, max_l, max_m):
    """
    Create an improved Physics-Informed Neural Network with residual connections
    and attention mechanisms
    """
    inputs = layers.Input(shape=input_shape)
    
    # Initial normalization
    x = layers.LayerNormalization()(inputs)
    
    # Enhanced convolutional backbone with residual connections
    def conv_block(x, filters, kernel_size):
        skip = x
        x = layers.Conv2D(filters, kernel_size, padding='same')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation('relu')(x)
        x = layers.Conv2D(filters, kernel_size, padding='same')(x)
        x = layers.BatchNormalization()(x)
        
        # Add skip connection if shapes match
        if skip.shape[-1] == filters:
            x = layers.Add()([x, skip])
        x = layers.Activation('relu')(x)
        return x
    
    # Multiple conv blocks with increasing complexity
    x = conv_block(x, 32, (3, 3))
    x = layers.MaxPooling2D((2, 2))(x)
    x = conv_block(x, 64, (3, 3))
    x = layers.MaxPooling2D((2, 2))(x)
    x = conv_block(x, 128, (3, 3))
    x = layers.MaxPooling2D((2, 2))(x)
    
    # Global attention mechanism
    attention = layers.GlobalAveragePooling2D()(x)
    attention = layers.Dense(128, activation='relu')(attention)
    attention = layers.Dense(128, activation='sigmoid')(attention)
    x = layers.Multiply()([x, layers.Reshape((1, 1, 128))(attention)])
    
    # Dense layers with dropout
    x = layers.Flatten()(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.4)(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    
    # Separate branches for n, l, m with shared features
    shared = layers.Dense(64, activation='relu')(x)
    
    # n branch
    n_branch = layers.Dense(32, activation='relu')(shared)
    n_output = layers.Dense(max_n, activation='softmax', name='n_output')(n_branch)
    
    # l branch (considers n_output)
    l_branch = layers.Concatenate()([shared, n_output])
    l_branch = layers.Dense(32, activation='relu')(l_branch)
    l_output = layers.Dense(max_l, activation='softmax', name='l_output')(l_branch)
    
    # m branch (considers both n_output and l_output)
    m_branch = layers.Concatenate()([shared, n_output, l_output])
    m_branch = layers.Dense(32, activation='relu')(m_branch)
    m_output = layers.Dense(max_m, activation='softmax', name='m_output')(m_branch)
    
    model = models.Model(inputs=inputs, outputs=[n_output, l_output, m_output])
    return model

def train_improved_pinn(train_dir, validation_split=0.2):
    """
    Train the improved PINN with advanced training strategies
    """
    # Load and preprocess data
    X_train, y_n, y_l, y_m = load_quantum_state_data(train_dir)
    
    # Advanced preprocessing
    X_train = X_train.astype('float32') / 255.0
    
    # Data augmentation
    data_augmentation = tf.keras.Sequential([
        layers.RandomRotation(0.1),
        layers.RandomZoom(0.1),
        layers.RandomTranslation(0.1, 0.1)
    ])
    
    # Get the number of classes
    max_n = np.max(y_n)
    max_l = np.max(y_l) + 1
    max_m = np.max(y_m) + 1
    
    # Create model
    model = create_improved_pinn_model(
        input_shape=(128, 128, 3),
        max_n=max_n,
        max_l=max_l,
        max_m=max_m
    )
    
    # Custom learning rate schedule
    initial_learning_rate = 0.001
    lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
        initial_learning_rate,
        decay_steps=1000,
        decay_rate=0.9,
        staircase=True
    )
    
    # Compile with custom loss
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=lr_schedule),
        loss={
            'n_output': lambda y_true, y_pred: advanced_physics_informed_loss(
                y_true, y_pred, 
                tf.argmax(model.get_layer('n_output').output, axis=-1),
                tf.argmax(model.get_layer('l_output').output, axis=-1),
                tf.argmax(model.get_layer('m_output').output, axis=-1)
            ),
            'l_output': lambda y_true, y_pred: advanced_physics_informed_loss(
                y_true, y_pred,
                tf.argmax(model.get_layer('n_output').output, axis=-1),
                tf.argmax(model.get_layer('l_output').output, axis=-1),
                tf.argmax(model.get_layer('m_output').output, axis=-1)
            ),
            'm_output': lambda y_true, y_pred: advanced_physics_informed_loss(
                y_true, y_pred,
                tf.argmax(model.get_layer('n_output').output, axis=-1),
                tf.argmax(model.get_layer('l_output').output, axis=-1),
                tf.argmax(model.get_layer('m_output').output, axis=-1)
            )
        },
        metrics=['accuracy']
    )
    
    # Convert labels to categorical
    y_n_cat = tf.keras.utils.to_categorical(y_n - 1, num_classes=max_n)
    y_l_cat = tf.keras.utils.to_categorical(y_l, num_classes=max_l)
    y_m_cat = tf.keras.utils.to_categorical(y_m, num_classes=max_m)
    
    # Train with advanced callbacks
    history = model.fit(
        data_augmentation(X_train),
        {
            'n_output': y_n_cat,
            'l_output': y_l_cat,
            'm_output': y_m_cat
        },
        epochs=300,
        batch_size=32,
        validation_split=validation_split,
        callbacks=[
            tf.keras.callbacks.EarlyStopping(
                monitor='val_loss',
                patience=30,
                restore_best_weights=True,
                min_delta=0.001
            ),
            tf.keras.callbacks.ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.7,
                patience=15,
                min_lr=1e-6
            ),
            tf.keras.callbacks.ModelCheckpoint(
                'best_model.h5',
                save_best_only=True,
                monitor='val_loss'
            )
        ]
    )
    
    return model, history

def plot_training_history(history):
    """
    Plot training history for all quantum numbers
    """
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    
    # Plot accuracy
    axes[0, 0].plot(history.history['n_output_accuracy'])
    axes[0, 0].plot(history.history['val_n_output_accuracy'])
    axes[0, 0].set_title('n Accuracy')
    axes[0, 0].legend(['Train', 'Validation'])
    
    axes[0, 1].plot(history.history['l_output_accuracy'])
    axes[0, 1].plot(history.history['val_l_output_accuracy'])
    axes[0, 1].set_title('l Accuracy')
    axes[0, 1].legend(['Train', 'Validation'])
    
    axes[0, 2].plot(history.history['m_output_accuracy'])
    axes[0, 2].plot(history.history['val_m_output_accuracy'])
    axes[0, 2].set_title('m Accuracy')
    axes[0, 2].legend(['Train', 'Validation'])
    
    # Plot loss
    axes[1, 0].plot(history.history['n_output_loss'])
    axes[1, 0].plot(history.history['val_n_output_loss'])
    axes[1, 0].set_title('n Loss')
    axes[1, 0].legend(['Train', 'Validation'])
    
    axes[1, 1].plot(history.history['l_output_loss'])
    axes[1, 1].plot(history.history['val_l_output_loss'])
    axes[1, 1].set_title('l Loss')
    axes[1, 1].legend(['Train', 'Validation'])
    
    axes[1, 2].plot(history.history['m_output_loss'])
    axes[1, 2].plot(history.history['val_m_output_loss'])
    axes[1, 2].set_title('m Loss')
    axes[1, 2].legend(['Train', 'Validation'])
    
    plt.tight_layout()
    plt.show()

def main():
    """
    Main function to run the training and testing
    """
    print("Training PINN model...")
    model, history = train_pinn()
    
    # Plot training history
    plot_training_history(history)
    
    # Test the model on new data
    test_dir = 'classification-data'
    X_test, y_n_test, y_l_test, y_m_test = load_quantum_state_data(test_dir)
    
    # Preprocess test data
    X_test = X_test.astype('float32') / 255.0
    
    # Make predictions
    test_predictions = model.predict(X_test)
    
    # Calculate and print accuracy
    n_pred = np.argmax(test_predictions[0], axis=1) + 1
    l_pred = np.argmax(test_predictions[1], axis=1)
    m_pred = np.argmax(test_predictions[2], axis=1)
    
    print("\nTest Results:")
    print("Predicted values:")
    print(f"n: {n_pred}")
    print(f"l: {l_pred}")
    print(f"m: {m_pred}")
    print("\nTrue values:")
    print(f"n: {y_n_test}")
    print(f"l: {y_l_test}")
    print(f"m: {y_m_test}")
    
    print("\nAccuracies:")
    print(f"n accuracy: {np.mean(n_pred == y_n_test):.4f}")
    print(f"l accuracy: {np.mean(l_pred == y_l_test):.4f}")
    print(f"m accuracy: {np.mean(m_pred == y_m_test):.4f}")

if __name__ == "__main__":
    main()
```
